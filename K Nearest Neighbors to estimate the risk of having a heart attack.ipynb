## Standardize the Variables

Because the KNN classifier predicts the class of a given test observation by identifying the observations that are nearest to it, the scale of the variables matters. Any variables that are on a large scale will have a much larger effect on the distance between the observations, and hence on the KNN classifier, than variables that are on a small scale.

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
data_to_scale= heart_data[["trtbps", "chol", "age", "thalachh","oldpeak" ]]
data_to_scale
```

```python
scaler.fit(data_to_scale)
```

```python
scaled_features = scaler.transform(data_to_scale)
```

```python
df_feat = pd.DataFrame(scaled_features,columns=data_to_scale.columns)
df_feat.head()
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
```

Concat all the features desired for the predictive model

```python
total_features= pd.concat([df_feat, heart_data[["sex", "fbs", "restecg", "exng","slp","caa","thall"]]], axis=1)
total_features
```

```python
X_train, X_test, y_train, y_test = train_test_split(total_features,heart_data['output'],
                                                    test_size=0.30)
```

## Predictions and Evaluations

To predict the best way possible the risk of having a heart attack, we need to find first de K value that minimize the error rate (type I and type II error)

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
error_rate = []

# Will take some time
for i in range(1,40):
    
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

A value of k equal to 7 seems the best choice to improve the performance of the predictive model

```python
knn = KNeighborsClassifier(n_neighbors=7)
```

```python
knn.fit(X_train,y_train)
```

```python
pred = knn.predict(X_test)
```

```python
print(confusion_matrix(y_test,pred))
```

```python
print(classification_report(y_test,pred))
```
